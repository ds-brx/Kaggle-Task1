# Task 1<br>
In theory, as you grow older, you have more chances to try new foods. Can this hypothesis be confirmed using the data in this dataset? Can it be graphed?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../input/have-you-ever-eaten-these-foods/responses.csv")

df = df.drop(["Race", "Home location", "Gender"], axis="columns")

pd.set_option('display.max_columns', None)
```

```python
df.info()
```

```python
df.head(27)
```

```python
# I'll assume that the NaN values are false and replacing them in the dataframe
df = df.fillna(0)
```

```python
df.isnull().sum()
```

```python
df.rename(columns={"Age in years": "age"},inplace=True)
```

```python
for col in df.columns:
    if col is not 'age':
        df[col] = df[col]*1
        result = df[col].astype(int)
        df[col] = result
```

```python
df.groupby('age').sum()
```

```python
for col in df.columns:
    if col is not 'age':
        print(df[col].value_counts())
```

```python
sum=0
for col in df.columns:
    if col is not 'age':
        sum = sum + df[col]
df['food'] = sum
```

```python
df['food'] = df['food']/df['food'].mean()
```

```python
df['food']
```

```python
plt.style.use('ggplot')
sns.lineplot(data=df, x="age", y="food")
```

The data seems to suggest that there is a tendency to consume more amount of food as you grow older.

```python
def plotByAge(col):
    fig, ax = plt.subplots()
    df.groupby(['age']).mean()[col].plot(ax=ax)
    ax.set_xlabel("age")
    ax.set_ylabel(col)
```

```python
for col in df.columns:
    if col is not 'age' and col is not 'food':
        plotByAge(col)
```

Observing from plots of various food items verses age groups, the tendency to explore new food increases with age.
